```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from matplotlib import cm

x=np.array([[2,0],[4,4],[6,2],[8,3]]) #입력변수: 학습시간, 과외시간 
y=np.array([81,93,91,97,98])

model=Sequential()

x1=np.array([2,4,6,8,7])
x2=np.array([0,4,2,3,4])

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter3D(x1,x2,y)
plt.show()

```

```python
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse') #확률적경사하강법, 평균오차제곱
model.fit(x,y,epochs=1000)

hour=7
private_class=4
prediction=model.predict([[hour,private_class]])

print("when We do study and private class during %.f hours and %.f hours then We can get %.02f grade point."%(hour,private_class,prediction))
```
